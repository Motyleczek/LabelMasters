```python
# 1. ogarnąć jak dobrze dawać tekst na zdjęcia

import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

np.random.seed = 420
```

```python
path_to_img = "generate_prep/blank_labels/blank_pocztex_1.png"

# cv2.destroyAllWindows()
# dajmy sobie adres na pocztexa, nadawca: 28,59 - 255,121

# co linijka:
# nazwa klienta (blank)
# Imie Nazwisko
# Adres (ulica plus numer)
# zip plus miasto
# kraj
# tel. (telefon)

nazwa_klienta = "ZOMO ENT"
imie_nazw = "Mateusz Zdrojowski"
adres = "Jana Wąsowicza 13/6"
zip_p_miasto = "33-044 Wałbrzych"
kraj = "POLSKA"
tel = "tel. +48654908878"
tekst_nadawcy = nazwa_klienta + "\n" + imie_nazw + "\n" + adres + "\n" + zip_p_miasto + "\n" + kraj+ "\n" + tel
print(tekst_nadawcy)
img = cv2.imread(path_to_img)
font = cv2.FONT_HERSHEY_SIMPLEX 
(x,y) = 28, 65

font_scale = 0.25
color = (0, 0, 0)
thickness = 1
font = cv2.FONT_HERSHEY_SIMPLEX
line_type = cv2.LINE_AA

text_size, _ = cv2.getTextSize(tekst_nadawcy, font, font_scale, thickness)
line_height = text_size[1] + 5
x, y0 = x, y
for i, line in enumerate(tekst_nadawcy.split("\n")):
    y = y0 + i * line_height
    cv2.putText(img,
                line,
                (x, y),
                font,
                font_scale,
                color,
                thickness,
                line_type)


plt.imshow(img)
plt.show()
```
